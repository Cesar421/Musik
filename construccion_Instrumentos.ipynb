```python
import math

def calcular_diseno_flauta(instrumento, nota_base, diametro_interno, espesor_pared, frecuencias_notas, diametros_agujeros):
    """
    Calcula la longitud del tubo y la posicion de los agujeros para instrumentos
    de viento andinos/tradicionales basados en las ecuaciones de Benade y Helmholtz.
    """
    # Constantes Fisicas
    v = 347.0  # Velocidad del sonido en m/s (ajustada a aire calido/aliento)
    R = (diametro_interno / 1000.0) / 2.0  # Radio interno en metros
    t = espesor_pared / 1000.0  # Espesor de pared en metros

    print("-" * 60)
    print(f"DISEÑO DE INSTRUMENTO: {instrumento.upper()} EN {nota_base.upper()}")
    print(f"Dimensiones del tubo: Diámetro Interno = {diametro_interno}mm, Espesor = {espesor_pared}mm")
    print("-" * 60)

    # 1. CALCULO DE LA LONGITUD TOTAL DEL TUBO
    f_base = frecuencias_notas[0]
    L_acustica_base = v / (2.0 * f_base)
    
    # Corrección de extremo final (universal)
    delta_L_extremo = 0.613 * R
    
    # Corrección de embocadura según el tipo de instrumento
    if instrumento.lower() == "traversa":
        delta_L_embocadura = 2.3 * R
    elif instrumento.lower() == "pincuyo":
        delta_L_embocadura = 1.5 * R
    elif instrumento.lower() == "quena":
        delta_L_embocadura = 0.1 * R  # Tiende a cero por estar totalmente abierto
    else:
        delta_L_embocadura = 0.0

    L_fisica_total = L_acustica_base - delta_L_embocadura - delta_L_extremo
    
    print(f"Longitud acústica teórica: {L_acustica_base * 100:.2f} cm")
    print(f"Longitud física real de corte: {L_fisica_total * 100:.2f} cm")
    print("Nota: Corta el tubo 1 o 2 cm más largo para ajustar en el taller.")
    print("-" * 60)
    print(f"{'Nota':<6}{'Freq (Hz)':<12}{'Diám Aguj.(mm)':<16}{'Dist. desde Boquilla (cm)':<25}")
    print("-" * 60)

    # 2. CALCULO DE LA POSICIÓN DE LOS AGUJEROS (De arriba hacia abajo)
    # Ignoramos la primera nota (f_base) ya que es el tubo completo abierto.
    # El bucle recorre las notas correspondientes a los 6 agujeros.
    posiciones_desde_boquilla = []
    
    for i in range(1, len(frecuencias_notas)):
        f_nota = frecuencias_notas[i]
        d_agujero = diametros_agujeros[i-1] # Diámetro de la broca en mm
        r = (d_agujero / 1000.0) / 2.0  # Radio del agujero en metros
        
        # Longitud acústica teórica requerida para esta nota específica
        L_acustica_nota = v / (2.0 * f_nota)
        
        # Ecuación de espesor efectivo y corrección del agujero (Benade / Calkin)
        t_efectivo = t + (1.5 * r)
        C_h = t_efectivo * ((R / r) ** 2)
        
        # Distancia física desde el "Punto Cero" (embocadura/ventana)
        D_x = L_acustica_nota - C_h
        posiciones_desde_boquilla.append(D_x)
        
        # Mostrar resultados formateados
        print(f"Aguj. {i}   {f_nota:<12.2f}{d_agujero:<16.1f}{D_x * 100:<25.2f}")

    print("-" * 60)
    print("DISTANCIAS CONSECUTIVAS ENTRE AGUJEROS (De centro a centro):")
    print("-" * 60)
    print(f"De Embocadura a Agujero 1: {posiciones_desde_boquilla[0]*100:.2f} cm")
    for idx in range(1, len(posiciones_desde_boquilla)):
        dist_intermedia = (posiciones_desde_boquilla[idx] - posiciones_desde_boquilla[idx-1]) * 100
        print(f"Del Agujero {idx} al Agujero {idx+1}: {dist_intermedia:.2f} cm")
    
    dist_al_final = (L_fisica_total - posiciones_desde_boquilla[-1]) * 100
    print(f"Del Agujero {len(posiciones_desde_boquilla)} al final del tubo: {dist_al_final:.2f} cm")
    print("-" * 60)


# =====================================================================
# CONFIGURACIÓN DEL EXPERIMENTO (Puedes modificar estos datos)
# =====================================================================

# EJEMPLO 1: Pincuyo en Re Mayor (Agudo) usando tubo de PVC delgado
tipo_instrumento = "pincuyo"
tonalidad = "Re"
dia_interno = 16.0   # en milímetros
esp_pared = 2.0      # en milímetros

# Frecuencias exactas de la escala diatónica de Re Mayor (Re5 a Do#6)
# [Re5 (base), Mi5, Fa#5, Sol5, La5, Si5, Do#6]
frecuencias_re = [587.33, 659.25, 739.99, 783.99, 880.00, 987.77, 1109.73]

# Diámetros de broca sugeridos (en mm) para cada uno de los 6 agujeros
diametros_brocas = [5.5, 6.0, 6.0, 5.5, 6.0, 5.5]

# Ejecutar el cálculo
calcular_diseno_flauta(tipo_instrumento, tonalidad, dia_interno, esp_pared, frecuencias_re, diametros_brocas)


# EJEMPLO 2: Quena en Sol Mayor (Profesional) usando tubo grueso (Descomenta para probar)
"""
tipo_instrumento = "quena"
tonalidad = "Sol"
dia_interno = 18.0
esp_pared = 3.0
# Escala de Sol Mayor: [Sol4 (base), La4, Si4, Do5, Re5, Mi5, Fa#5]
frecuencias_sol = [392.00, 440.00, 493.88, 523.25, 587.33, 659.25, 739.99]
diametros_brocas_quena = [7.0, 7.5, 6.5, 6.0, 7.5, 5.5]

calcular_diseno_flauta(tipo_instrumento, tonalidad, dia_interno, esp_pared, frecuencias_sol, diametros_brocas_quena)
"""
```

```python
import math

def calcular_diseno_flauta_referenciado(instrumento, nota_base, diametro_interno, espesor_pared, nombres_notas, frecuencias_notas, diametros_agujeros):
    """
    Calcula la longitud del tubo y la posición de los agujeros referenciados 
    ESTRICTAMENTE al punto de embocadura (donde se sopla) usando Benade y Helmholtz.
    """
    # Constantes Físicas
    v = 347.0  # Velocidad del sonido en m/s (ajustada al aliento cálido interno)
    R = (diametro_interno / 1000.0) / 2.0  # Radio interno en metros
    t = espesor_pared / 1000.0  # Espesor de pared en metros

    print("=" * 75)
    print(f"DISEÑO ACÚSTICO: {instrumento.upper()} EN {nota_base.upper()}")
    print(f"Tubo: Diámetro Interno = {diametro_interno}mm | Espesor de Pared = {espesor_pared}mm")
    print("=" * 75)
    print("⚠️ TODAS LAS MEDIDAS ABAJO ESTÁN REFERENCIADAS AL CENTRO DE LA EMBOCADURA (PUNTO 0)")
    print("=" * 75)

    # 1. CÁLCULO DE LA LONGITUD DESDE LA EMBOCADURA HASTA EL FINAL DEL TUBO
    f_base = frecuencias_notas[0]
    L_acustica_base = v / (2.0 * f_base)
    
    # Correcciones de extremo
    delta_L_extremo = 0.613 * R
    
    if instrumento.lower() == "traversa":
        delta_L_embocadura = 2.3 * R
    elif instrumento.lower() == "pincuyo":
        delta_L_embocadura = 1.5 * R
    elif instrumento.lower() == "quena":
        delta_L_embocadura = 0.1 * R
    else:
        delta_L_embocadura = 0.0

    # Longitud física desde el centro de la embocadura hasta el final abierto del tubo
    L_desde_embocadura_al_final = L_acustica_base - delta_L_embocadura - delta_L_extremo
    
    # Nota de taller sobre el tapón trasero (solo para traversa y pincuyo)
    if instrumento.lower() in ["traversa", "pincuyo"]:
        print(f"-> Extremo trasero (Tapón/Corcho): Debe ir colocado a {-delta_L_embocadura * 100:.2f} cm hacia atrás del punto 0.")
    print(f"-> Extremo delantero (Final del tubo): Debe cortarse a {L_desde_embocadura_al_final * 100:.2f} cm desde el punto 0.")
    print("-" * 75)
    
    # Encabezado de la tabla
    print(f"{'Elemento':<12}{'Nota al Destapar':<20}{'Freq (Hz)':<12}{'Diám. Broca':<15}{'Dist. desde Embocadura'}")
    print("-" * 75)

    # 2. CÁLCULO DE CADA HUECO (Referenciado al Punto Cero de soplado)
    # Recorremos desde el agujero 1 (el más cercano a la boca) al 6 (el más lejano)
    posiciones_desde_embocadura = []
    
    for i in range(1, len(frecuencias_notas)):
        f_nota = frecuencias_notas[i]
        nota_nombre = nombres_notas[i]
        d_agujero = diametros_agujeros[i-1] 
        r = (d_agujero / 1000.0) / 2.0  
        
        # Longitud acústica teórica para la nota del hueco
        L_acustica_nota = v / (2.0 * f_nota)
        
        # Corrección del agujero (Ecuación de Benade / Calkin)
        t_efectivo = t + (1.5 * r)
        C_h = t_efectivo * ((R / r) ** 2)
        
        # Distancia física real medida desde el centro de la embocadura
        D_x = L_acustica_nota - C_h
        posiciones_desde_embocadura.append(D_x)
        
        print(f"Hueco {i:<5}{nota_nombre:<20}{f_nota:<12.2f}{d_agujero:<15.1f}{D_x * 100:.2f} cm")

    # Añadir el extremo final a la lista para el desglose visual
    print(f"Final tubo  {nombres_notas[0]:<20}{f_base:<12.2f}{'-':<15}{L_desde_embocadura_al_final * 100:.2f} cm")
    print("=" * 75)


# =====================================================================
# CONFIGURACIÓN DE TU EXPERIMENTO EN EL TALLER
# =====================================================================

# EJEMPLO: Pincuyo / Flauta en Re Mayor (Aguda)
# Si vas a hacer una Flauta Traversa cambia "pincuyo" por "traversa"
tipo_instrumento = "pincuyo"  
nota_fundamental = "Re (D)"
dia_interno = 16.0   # mm
esp_pared = 2.0      # mm

# Nombres de las notas de la escala para cada estado de la flauta
# Índices: [0] es la nota base (todos tapados). Del [1] al [6] es la nota que suena AL DESTAPAR ese hueco.
nombres_escala_re = [
    "Re (Todos tapados)", 
    "Do# (Destapar Hueco 1)", 
    "Si  (Destapar Hueco 2)", 
    "La  (Destapar Hueco 3)", 
    "Sol (Destapar Hueco 4)", 
    "Fa# (Destapar Hueco 5)", 
    "Mi  (Destapar Hueco 6)"
]

frecuencias_escala_re = [587.33, 1109.73, 987.77, 880.00, 783.99, 739.99, 659.25]
diametros_brocas = [5.5, 6.0, 6.0, 5.5, 6.0, 5.5]

# Ejecutar cálculo indexado al punto de soplado
calcular_diseno_flauta_referenciado(
    tipo_instrumento, 
    nota_fundamental, 
    dia_interno, 
    esp_pared, 
    nombres_escala_re, 
    frecuencias_escala_re, 
    diametros_brocas
)
```

```python
import math

def calcular_diseno_flauta_completo(instrumento, tonalidad, dia_interno, espesor_pared, material, nombres_notas, frecuencias, diametros_brocas):
    """
    Calcula las especificaciones técnicas para construir instrumentos de viento.
    Basado en el modelo de impedancia acústica de Benade/Helmholtz.
    """
    # Constantes físicas
    v = 347.0  # Velocidad del sonido en m/s (ajustada a temperatura de aliento)
    R = (dia_interno / 1000.0) / 2.0  # Radio interno en metros
    t = espesor_pared / 1000.0        # Espesor en metros

    # Cálculo de longitud física (Benade/Helmholtz)
    L_acustica_base = v / (2.0 * frecuencias[0])
    delta_L_extremo = 0.613 * R
    
    # Corrección de embocadura según tipo de instrumento
    if instrumento.lower() == "traversa":
        delta_L_embocadura = 2.3 * R
    elif instrumento.lower() == "pincuyo":
        delta_L_embocadura = 1.5 * R
    elif instrumento.lower() == "quena":
        delta_L_embocadura = 0.1 * R
    else:
        delta_L_embocadura = 0.0

    L_fisica = L_acustica_base - delta_L_embocadura - delta_L_extremo

    # Impresión del informe técnico
    print("="*80)
    print(f"INFORME TÉCNICO DE CONSTRUCCIÓN: {instrumento.upper()} | {tonalidad}")
    print("="*80)
    print(f"1. ESPECIFICACIONES DE MATERIAL:")
    print(f"   - Material recomendado: {material}")
    print(f"   - Diámetro Interno: {dia_interno} mm")
    print(f"   - Espesor de pared: {espesor_pared} mm")
    print(f"\n2. DIMENSIONES GLOBALES:")
    print(f"   - Longitud física total: {L_fisica * 100:.2f} cm")
    print(f"   - Nota base (todos los huecos cerrados): {nombres_notas[0]}")
    
    print("\n3. GUÍA DE PERFORACIÓN (Referencia desde el centro de la embocadura - Punto 0):")
    print(f"{'Elemento':<15} | {'Nota':<12} | {'Diám. (mm)':<12} | {'Posición (cm)':<10}")
    print("-"*70)
    
    for i in range(1, len(frecuencias)):
        r = (diametros_brocas[i-1] / 1000.0) / 2.0
        t_efec = t + (1.5 * r)
        C_h = t_efec * ((R / r) ** 2)
        distancia = (v / (2.0 * frecuencias[i]) - C_h) * 100
        print(f"Hueco {i:<9} | {nombres_notas[i]:<12} | {diametros_brocas[i-1]:<12.1f} | {distancia:.2f} cm")
    
    print("-"*70)
    print(f"Extremo final   | Final tubo   | ---          | {L_fisica * 100:.2f} cm")
    print("="*80)
    print("NOTAS PARA EL CONSTRUCTOR:")
    print("- Punto 0: Centro del orificio de soplado (embocadura/ventana).")
    print("- Si la nota suena baja (bemol): agranda el hueco 0.5mm y prueba de nuevo.")
    print("- Si la nota suena alta (sostenido): el hueco está muy cerca de la boquilla (ajusta el tapón).")
    print("="*80)

# --- EJECUCIÓN DEL CÓDIGO ---

# Definición de parámetros para Pincuyo en Re (puedes cambiar los valores aquí)
calcular_diseno_flauta_completo(
    instrumento="pincuyo",
    tonalidad="Re Mayor",
    dia_interno=16.0,
    espesor_pared=2.0,
    material="PVC de Alta Presión o Bambú",
    # Notas: [0] es la base, [1-6] son los huecos al destapar
    nombres_notas=["Re5", "Do#6", "Si5", "La5", "Sol5", "Fa#5", "Mi5"],
    frecuencias=[587.33, 1109.73, 987.77, 880.00, 783.99, 739.99, 659.25],
    diametros_brocas=[5.5, 6.0, 6.0, 5.5, 6.0, 5.5]
)
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
============================================================================
 TALLER DE QUENAS & PINCUYOS  ·  Diseñador acústico de flautas andinas
============================================================================
Calcula el largo del tubo y la posición exacta de cada agujero para construir
quenas, pincuyos y flautas traversas en cualquier afinación.

Modelo de impedancia acústica de Benade / Helmholtz (tubo abierto), con:
  - generación de frecuencias por temperamento igual (A4 = 440 Hz),
  - velocidad del sonido en función de la temperatura del aire,
  - corrección de embocadura según el tipo de instrumento,
  - validaciones de geometría.

Sin dependencias externas: solo necesita Python 3.
Uso:
    python3 taller_de_quenas.py          -> menú interactivo
    o importar las funciones desde otro script.
============================================================================
"""

import math
from dataclasses import dataclass, field
from typing import List

# ---------------------------------------------------------------------------
# 1. TEORÍA MUSICAL  ·  notas y frecuencias (temperamento igual)
# ---------------------------------------------------------------------------

NOTAS_ES = ["Do", "Do#", "Re", "Re#", "Mi", "Fa",
            "Fa#", "Sol", "Sol#", "La", "La#", "Si"]
NOTAS_EN = ["C", "C#", "D", "D#", "E", "F",
            "F#", "G", "G#", "A", "A#", "B"]

# Grados (en semitonos) de cada escala respecto a la nota base
ESCALAS = {
    "mayor": [0, 2, 4, 5, 7, 9, 11],
    "menor": [0, 2, 3, 5, 7, 8, 10],   # menor natural
}

# Corrección de embocadura como múltiplo del radio interno R
EMBOCADURA = {
    "quena":    0.1,   # extremo totalmente abierto (tiende a 0)
    "pincuyo":  1.5,   # pico / fipple
    "traversa": 2.3,   # transversal con extremo tapado
}


def indice_nota(nombre: str) -> int:
    """Devuelve el índice 0-11 de una nota a partir de su nombre (ES o EN)."""
    n = nombre.strip().capitalize()
    if n in NOTAS_ES:
        return NOTAS_ES.index(n)
    if n.upper() in NOTAS_EN:
        return NOTAS_EN.index(n.upper())
    raise ValueError(f"Nota desconocida: {nombre!r}")


def nombre_desde_midi(midi: int) -> str:
    """Nombre en español de una nota MIDI, p.ej. 'Sol'."""
    return NOTAS_ES[midi % 12]


def midi(indice: int, octava: int) -> int:
    """Número MIDI a partir del índice de nota y la octava (Do4 = 60)."""
    return 12 * (octava + 1) + indice


def frecuencia(midi_num: int) -> float:
    """Frecuencia en Hz por temperamento igual (A4 = 440 Hz)."""
    return 440.0 * (2.0 ** ((midi_num - 69) / 12.0))


def velocidad_sonido(temp_c: float) -> float:
    """Velocidad del sonido en aire (m/s) a una temperatura dada en °C."""
    return 331.3 * math.sqrt(1.0 + temp_c / 273.15)


# ---------------------------------------------------------------------------
# 2. MODELO DE DATOS
# ---------------------------------------------------------------------------

@dataclass
class Agujero:
    n: int            # número de agujero (1 = más cercano a la boca)
    nombre: str       # nota que suena al destaparlo
    freq: float       # Hz
    diametro: float   # mm
    pos_cm: float     # distancia desde el Punto 0 (centro de la embocadura)
    tramo_cm: float   # distancia de centro a centro con el agujero anterior


@dataclass
class Diseno:
    instrumento: str
    nota_base: str        # p.ej. "Sol4"
    escala: str
    dia_interno: float    # mm
    espesor: float        # mm
    temp_c: float
    v_sonido: float       # m/s
    freq_base: float      # Hz
    largo_cm: float       # de la boca al final del tubo
    tapon_cm: float       # distancia del tapón detrás del Punto 0 (0 en quena)
    agujeros: List[Agujero] = field(default_factory=list)
    avisos: List[str] = field(default_factory=list)


# ---------------------------------------------------------------------------
# 3. CÁLCULO ACÚSTICO PRINCIPAL  (Benade / Helmholtz)
# ---------------------------------------------------------------------------

def calcular_diseno(instrumento: str,
                    nota: str,
                    octava: int,
                    escala: str = "mayor",
                    dia_interno: float = 18.0,
                    espesor_pared: float = 3.0,
                    diametros_agujeros: List[float] = None,
                    temp_c: float = 27.0) -> Diseno:
    """
    Calcula toda la geometría del instrumento.

    Parámetros
    ----------
    instrumento : 'quena' | 'pincuyo' | 'traversa'
    nota        : nota base con todos los agujeros tapados (p.ej. 'Sol' o 'G')
    octava      : octava de la nota base (Do4 = octava media)
    escala      : 'mayor' o 'menor'
    dia_interno : diámetro interno del tubo, en mm
    espesor_pared : espesor de la pared, en mm
    diametros_agujeros : lista de 6 diámetros (mm), de la boca hacia la punta.
                         Si es None se usan 7 mm por defecto.
    temp_c      : temperatura del aire dentro del tubo, en °C
    """
    instrumento = instrumento.lower()
    if instrumento not in EMBOCADURA:
        raise ValueError(f"Instrumento no soportado: {instrumento!r}")
    if escala not in ESCALAS:
        raise ValueError(f"Escala no soportada: {escala!r}")
    if diametros_agujeros is None:
        diametros_agujeros = [7.0] * 6
    if len(diametros_agujeros) != 6:
        raise ValueError("Se requieren exactamente 6 diámetros de agujero.")

    # --- constantes físicas ---
    v = velocidad_sonido(temp_c)
    R = (dia_interno / 1000.0) / 2.0     # radio interno (m)
    t = espesor_pared / 1000.0           # espesor (m)

    # --- nota base y longitud total del tubo ---
    base_idx = indice_nota(nota)
    base_midi = midi(base_idx, octava)
    f_base = frecuencia(base_midi)

    L_acustica = v / (2.0 * f_base)
    d_extremo = 0.613 * R                       # corrección del extremo abierto
    d_embocadura = EMBOCADURA[instrumento] * R  # corrección de embocadura
    L_fisica = L_acustica - d_embocadura - d_extremo

    # --- posición de cada agujero (grados 2º…7º de la escala) ---
    offsets = ESCALAS[escala]
    crudos = []
    for i in range(1, len(offsets)):                 # 6 agujeros
        m = base_midi + offsets[i]
        f = frecuencia(m)
        d = diametros_agujeros[i - 1]
        r = (d / 1000.0) / 2.0
        t_efectivo = t + 1.5 * r                     # espesor efectivo (Benade)
        C_h = t_efectivo * ((R / r) ** 2)            # corrección del agujero
        pos = v / (2.0 * f) - C_h                    # distancia desde el Punto 0
        crudos.append((m, f, d, pos))

    # el agujero más agudo queda más cerca de la boca
    crudos.sort(key=lambda x: x[3])

    # --- construir lista de agujeros con tramos centro a centro ---
    agujeros: List[Agujero] = []
    prev = 0.0
    for i, (m, f, d, pos) in enumerate(crudos):
        pos_cm = pos * 100.0
        tramo = pos_cm if i == 0 else pos_cm - prev * 100.0
        agujeros.append(Agujero(i + 1, nombre_desde_midi(m), f, d, pos_cm, tramo))
        prev = pos

    # --- validaciones de geometría ---
    avisos: List[str] = []
    if L_fisica <= 0:
        avisos.append("Las correcciones superan el largo del tubo: "
                      "el Ø interno es demasiado grande para esa nota.")
    anterior = 0.0
    for a in agujeros:
        if a.pos_cm <= 0:
            avisos.append(f"Agujero {a.n} ({a.nombre}): posición imposible; "
                          "aumenta su Ø o revisa el tubo.")
        if a.n > 1 and a.pos_cm <= anterior:
            avisos.append(f"Agujeros {a.n - 1} y {a.n} se solapan; ajusta sus diámetros.")
        anterior = a.pos_cm
    if agujeros and agujeros[-1].pos_cm >= L_fisica * 100.0:
        avisos.append("El último agujero cae fuera del tubo; revisa afinación o dimensiones.")

    return Diseno(
        instrumento=instrumento,
        nota_base=f"{nombre_desde_midi(base_midi)}{octava}",
        escala=escala,
        dia_interno=dia_interno,
        espesor=espesor_pared,
        temp_c=temp_c,
        v_sonido=v,
        freq_base=f_base,
        largo_cm=L_fisica * 100.0,
        tapon_cm=(d_embocadura * 100.0) if instrumento != "quena" else 0.0,
        agujeros=agujeros,
        avisos=avisos,
    )


# ---------------------------------------------------------------------------
# 4. INFORME EN TEXTO
# ---------------------------------------------------------------------------

def imprimir_informe(d: Diseno) -> None:
    """Imprime un informe de taller legible y un esquema ASCII a escala."""
    print("=" * 74)
    print(f" INFORME DE CONSTRUCCIÓN  ·  {d.instrumento.upper()} en {d.nota_base} {d.escala}")
    print("=" * 74)
    print(f" Ø interno: {d.dia_interno} mm   |   Espesor pared: {d.espesor} mm")
    print(f" Temperatura: {d.temp_c} °C   ->   velocidad del sonido = {d.v_sonido:.1f} m/s")
    print(f" Nota base (todo tapado): {d.nota_base}  ({d.freq_base:.1f} Hz)")
    print("-" * 74)
    print(f" LARGO DE LA BOCA AL FINAL DEL TUBO: {d.largo_cm:.2f} cm")
    print(" (corta el tubo 1-2 cm más largo y afina recortando)")
    if d.instrumento != "quena":
        print(f" Tapón / corcho trasero: {d.tapon_cm:.2f} cm DETRÁS del Punto 0")
    else:
        print(" Punto 0 = filo de la escotadura por donde soplas (extremo abierto)")
    print("-" * 74)

    print(f" {'Aguj':<5}{'Nota':<6}{'Frec(Hz)':>10}{'Ø(mm)':>8}"
          f"{'Desde boca':>14}{'Tramo':>10}")
    print("-" * 74)
    for a in d.agujeros:
        print(f" {a.n:<5}{a.nombre:<6}{a.freq:>10.1f}{a.diametro:>8.1f}"
              f"{a.pos_cm:>12.2f}cm{a.tramo_cm:>8.2f}cm")
    tramo_final = d.largo_cm - d.agujeros[-1].pos_cm
    print(f" {'—':<5}{'Final':<6}{d.freq_base:>10.1f}{'—':>8}"
          f"{d.largo_cm:>12.2f}cm{tramo_final:>8.2f}cm")
    print("=" * 74)

    # esquema ASCII a escala (60 caracteres = largo del tubo)
    _esquema_ascii(d)

    # avisos / consejos
    if d.avisos:
        print("\n ⚠ AVISOS:")
        for w in d.avisos:
            print(f"   - {w}")
    print("\n CONSEJOS DE TALLER:")
    print("   - Nota baja (bemol)  -> agranda ese agujero ~0,5 mm.")
    print("   - Nota alta (sostenido) -> el agujero está muy cerca de la boca.")
    print("   - Perfora con broca pequeña y agranda escuchando un afinador.")
    print("   - Todas las medidas se toman desde el Punto 0 (centro de la embocadura).")
    print("=" * 74 + "\n")


def _esquema_ascii(d: Diseno, ancho: int = 60) -> None:
    """Dibuja una regla ASCII con la posición de los agujeros."""
    fila = ["-"] * (ancho + 1)
    fila[0] = "O"  # embocadura
    for a in d.agujeros:
        x = int(round(a.pos_cm / d.largo_cm * ancho))
        x = min(max(x, 0), ancho)
        fila[x] = str(a.n)
    print(" Esquema (boca [O] -> punta):")
    print("   " + "".join(fila) + "|")
    print(f"   0 cm{' ' * (ancho - 8)}{d.largo_cm:.1f} cm")


# ---------------------------------------------------------------------------
# 5. MENÚ INTERACTIVO
# ---------------------------------------------------------------------------

def _preguntar(texto, defecto, conv=str):
    """input() con valor por defecto al pulsar Enter."""
    r = input(f"{texto} [{defecto}]: ").strip()
    if r == "":
        return defecto
    try:
        return conv(r)
    except ValueError:
        print("   (entrada inválida, uso el valor por defecto)")
        return defecto


# Dimensiones típicas por instrumento [Ø interno, pared, Ø agujeros boca->punta]
PRESETS = {
    "quena":    dict(dia=18.0, pared=3.0, holes=[8, 8, 7.5, 7, 7, 6.5],
                     nota="Sol", oct=4),
    "pincuyo":  dict(dia=16.0, pared=2.0, holes=[6, 6, 6, 5.5, 5.5, 5.5],
                     nota="Re", oct=5),
    "traversa": dict(dia=19.0, pared=3.0, holes=[7, 7, 7, 6.5, 6.5, 6],
                     nota="Re", oct=4),
}


def menu():
    print("\n  TALLER DE QUENAS & PINCUYOS  ·  diseñador acústico\n")
    instr = _preguntar("Instrumento (quena/pincuyo/traversa)", "quena").lower()
    if instr not in PRESETS:
        print("   (no reconocido, uso 'quena')")
        instr = "quena"
    p = PRESETS[instr]

    nota = _preguntar("Nota base (Do, Re, Mi, ... o C, D, E...)", p["nota"])
    octava = _preguntar("Octava", p["oct"], int)
    escala = _preguntar("Escala (mayor/menor)", "mayor").lower()
    dia = _preguntar("Ø interno (mm)", p["dia"], float)
    pared = _preguntar("Espesor de pared (mm)", p["pared"], float)
    temp = _preguntar("Temperatura del aire (°C)", 27.0, float)

    usar_def = _preguntar("¿Usar Ø de agujeros por defecto? (s/n)", "s").lower()
    if usar_def.startswith("s"):
        holes = [float(x) for x in p["holes"]]
    else:
        holes = []
        for i in range(6):
            holes.append(_preguntar(f"  Ø agujero {i + 1} (mm)",
                                    float(p["holes"][i]), float))

    diseno = calcular_diseno(instr, nota, octava, escala, dia, pared, holes, temp)
    print()
    imprimir_informe(diseno)


# ---------------------------------------------------------------------------
# 6. EJEMPLOS LISTOS PARA CORRER
# ---------------------------------------------------------------------------

def ejemplos():
    """Tres diseños de muestra para ver el formato de salida."""
    # Quena profesional en Sol mayor
    imprimir_informe(calcular_diseno(
        "quena", "Sol", 4, "mayor",
        dia_interno=18.0, espesor_pared=3.0,
        diametros_agujeros=[8, 8, 7.5, 7, 7, 6.5], temp_c=27.0))

    # Pincuyo agudo en Re mayor
    imprimir_informe(calcular_diseno(
        "pincuyo", "Re", 5, "mayor",
        dia_interno=16.0, espesor_pared=2.0,
        diametros_agujeros=[6, 6, 6, 5.5, 5.5, 5.5], temp_c=27.0))

    # Traversa en Re mayor
    imprimir_informe(calcular_diseno(
        "traversa", "Re", 4, "mayor",
        dia_interno=19.0, espesor_pared=3.0,
        diametros_agujeros=[7, 7, 7, 6.5, 6.5, 6], temp_c=27.0))


if __name__ == "__main__":
    # Por defecto abre el menú interactivo.
    # Para ver los ejemplos sin interacción, cambia a:  ejemplos()
    try:
        menu()
    except (KeyboardInterrupt, EOFError):
        print("\nHasta la próxima. 🎵")
```
